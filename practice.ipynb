```python
import pandas as pd
import matplotlib.pyplot as plt

# Example DataFrame
df = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [4, 5, 6, 7],
    'C': [7, 8, 9, 10]
})

# Compute means
means = df.mean()

# Create scatter plot
plt.figure(figsize=(6, 4))
plt.scatter(means.index, means.values, s=100, color='red')  # s controls dot size
plt.ylabel("Mean")
plt.xlabel("Columns")
plt.title("Mean of Each Column")

# Optional: annotate points with their values
for i, val in enumerate(means.values):
    plt.text(means.index[i], val, f"{val:.2f}", ha='center', va='bottom')

plt.show()
```

```python
def stem_and_leaf(data):
    data = sorted(int(x) for x in data)  # sort and ensure integers
    stems = {}

    for num in data:
        stem = num // 10       # tens place
        leaf = num % 10        # ones place
        stems.setdefault(stem, []).append(leaf)

    # Print result
    for stem, leaves in stems.items():
        leaf_str = " ".join(str(l) for l in leaves)
        print(f"{stem} | {leaf_str}")


# Example
data = [12, 15, 22, 23, 24, 35, 37, 42, 43, 44, 45, 48]
stem_and_leaf(data)
```

```python
from PIL import Image

def split_into_squares(image_path, rows, cols, save=True):
    img = Image.open(image_path)
    w, h = img.size

    tile_w = w // cols
    tile_h = h // rows

    squares = []

    for r in range(rows):
        for c in range(cols):
            left = c * tile_w
            upper = r * tile_h
            right = left + tile_w
            lower = upper + tile_h

            tile = img.crop((left, upper, right, lower))
            squares.append(tile)

            if save:
                tile.save(f"tile_{r}_{c}.png")

    return squares
```

```python
split_into_squares('crowd_counting.jpg', 2, 2)
```

```python
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open("crowd.png")
w, h = img.size

rows, cols = 20, 20

fig, ax = plt.subplots()
ax.imshow(img)
ax.axis("off")

# Horizontal lines
for i in range(1, rows):
    ax.axhline(i * h / rows, linewidth=1, color="white")

# Vertical lines
for j in range(1, cols):
    ax.axvline(j * w / cols, linewidth=1, color="white")

plt.show()
```

```python
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open("counting1.jpg")
w, h = img.size

rows, cols = 20, 20

fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(img)
ax.axis("off")

# Horizontal lines
for i in range(1, rows):
    ax.axhline(i * h / rows, linewidth=0.3, color="white")

# Vertical lines
for j in range(1, cols):
    ax.axvline(j * w / cols, linewidth=0.3, color="white")

#  Save full image (no cropping)
plt.savefig(
    "counting1_grid.png",
    dpi=300,
    bbox_inches="tight",
    pad_inches=0
)

plt.show()
```
